```python
import pandas as pd
```

```python
participants=['01','02','03','04','05','06','07','08','09']
for i in range(10,27):
    participants.append(str(i))
print(participants)
```

```python
combined_df=pd.DataFrame()
for p in participants:

    df = pd.read_csv('D:\FYP\Datasets\dataset2_features\oxy\skew_'+p+'.csv')
    combined_df = pd.concat([combined_df,df])
combined_df=combined_df.drop(['Unnamed: 0'],axis=1)
print(combined_df)
combined_df.to_csv('D:\FYP\Datasets\dataset2_features\oxy\skew.csv')
```
