```python
import scipy.io as sio
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
```

```python
participants=['01','02','03','04','05','06','07','08','09']
for i in range(10,27):
    participants.append(str(i))
print(participants)
```

```python
from scipy.stats import skew
from scipy.stats import entropy
```

```python
for participant in participants:

    mat_data = sio.loadmat('D:\FYP\Datasets\dataset2\VP0'+participant+'-NIRS\mrk_nback')
    my_data = mat_data['mrk_nback']['time'] # For a structure
    #print(my_data[0][0][0])
    time=my_data[0][0][0]
    df=pd.read_csv('D:\FYP\Datasets\dataset2_csv\deoxy\deoxy_'+participant+'.csv')
    new=np.array(time)
    time_new=np.insert(new,0,0)
    #print(time_new)
    numDataPoints = np.diff(time_new) / 100  # Sampling rate is 10 Hz (adjust if needed)
    numDataPoints = numDataPoints.astype(int) 
    #print(numDataPoints)
    val=[]
    start_index=0
    for t in numDataPoints:
        #print('session')
        end_index=start_index+t
        #print('start and end',start_index,end_index)
        newdf=df.iloc[start_index:end_index,:]
        start_index=end_index
        mean_values=[]
        for channel in range(36):
            x=newdf.iloc[:,channel]
            #print(len(x))
            #print(channel)
            y=savgol_filter(x, 121, 3)
            baseline_mean=np.mean(y)
            y=y-baseline_mean
            slope, _ = np.polyfit(np.arange(len(y)), y, 1)
            mean_values.append(slope)
            #print(np.mean(y))
        #print(mean_values)
        val.append(mean_values)
    meandf=pd.DataFrame(val)
    #print(meandf)
    filename='D:\FYP\Datasets\dataset2_features\deoxy\slope_'+participant+'.csv'
    meandf.to_csv(filename)
```
